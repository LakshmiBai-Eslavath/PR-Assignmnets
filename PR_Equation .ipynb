```python
import numpy as np
```

```python
in_matrix=[]
r=int(input("enter no.of rows: "))
c=int(input("enter no.of columns: "))
T_matrix,l_matrix,R_matrix,RR_result,RRR_matrix,final_matrix=[],[],[],[],[],[]

#dynamic memory allocation
def my_function(r1,c1):
  l_matrix=[]
  for i in range(c1):
    temp=[]
    for j in range(r1):
        temp.append(int(0))
    l_matrix.append(temp)
  return l_matrix

# input for matrix Y as in_matrix
print("enter values row wise:")
for i in range(r):
    temp=[]
    for j in range(c):
        temp.append(int(input()))
    in_matrix.append(temp)
print("Y:",in_matrix)

T_matrix=my_function(r,c)
#print(T_matrix)

# transpose of matrix
for i in range(r):
    for j in range(c):
        T_matrix[j][i]=in_matrix[i][j]
print("Y^T:",T_matrix)

R_matrix=my_function(len(T_matrix),len(in_matrix[0]))
#print(R_matrix)

# mul of transpose matrix and input matrix
for i in range(len(T_matrix)):
    for j in range(len(in_matrix[0])):
        for k in range(len(in_matrix)):
            R_matrix[i][j]+=T_matrix[i][k]*in_matrix[k][j]
print("Y^TY:",R_matrix)

RR_result=my_function(len(R_matrix),len(R_matrix[0]))


# inverse of matrix
RR_result=np.linalg.inv(R_matrix)
print("(Y^TY)^(-1):",RR_result)

#print(len(RR_result))
#print(len(T_matrix[0]))
RRR_matrix=my_function(len(T_matrix[0]),len(RR_result))
#print(RRR_matrix)

# mul of RR_result and T_matrix
for i in range(len(RR_result)):
    for j in range(len(T_matrix[0])):
        for k in range(len(T_matrix)):
            RRR_matrix[i][j]+=RR_result[i][k]*T_matrix[k][j]
            
print("(Y^TY)^(-1)*Y^T:",RRR_matrix)

# input for B matrix
b_matrix=[]
b_r=int(input("enter no.of rows for b matrix: "))
b_c=int(input("enter no.of columns for b matrix: "))
print("enter values row wise:")
for i in range(b_r):
    temp1=[]
    for j in range(b_c):
        temp1.append(int(input()))
    b_matrix.append(temp1)
print("b matrix: ",b_matrix)
#print(len(b_matrix[0]))
#print(len(RRR_matrix))

final_matrix=my_function(len(b_matrix[0]),len(RRR_matrix))

#print(final_result)
#print(final_matrix)
#print(len(b_matrix[0]))
#print(len(b_matrix))

# mul of RRR_matrix with b matrix
if len(RRR_matrix[0])==b_r:
    for i in range(len(RRR_matrix)):
        for j in range(b_c):
            for k in range(b_r):
                final_matrix[i][j]+=RRR_matrix[i][k]*b_matrix[k][j]
    print("A Value: ",final_matrix)
else:
    print("multiplication not possible")
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.plot(in_matrix,b_matrix, "b.")
plt.plot(in_matrix,b_matrix, "r-", label="Predictions")
plt.xlabel("$x_1$", fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 4, 0, 8])
plt.show()
```
